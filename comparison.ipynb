```python
import pickle
import statistics
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

# Charger les données du premier fichier pickle
it = 10
with open(f"/home/stagiaire/D/R/metrics/MA{it}3.pkl", 'rb') as f:
    global_list1 = pickle.load(f)

# Préparer les données pour le premier ensemble
data1 = {}
for i in range(1, 5):
    root_dir = f"/home/stagiaire/D/D/patchs/{i}"
    keys = {"favelas": 1, "residential": 0}
    for folder in os.listdir(root_dir):
        root_folder = os.path.join(root_dir, folder)
        files = os.listdir(root_folder)
        for file in files:
            n = file.split('.')[0]
            data1[n] = keys[folder]
labels_list1 = pd.DataFrame(list(data1.items()), columns=['Id', 'Label'])
labels_list1 = labels_list1.set_index('Id')
global_list1 = pd.merge(global_list1, labels_list1, left_index=True, right_index=True)

for index, row in global_list1.iterrows():
    if not row['PredictC']:
        global_list1 = global_list1.drop(index)
global_list1['medianC'] = (global_list1['PredictC'].apply(statistics.median)).apply(lambda m: 1 if m == 0.5 else m)
global_list1['medianA'] = (global_list1['PredictA'].apply(statistics.median)).apply(lambda m: 1 if m == 0.5 else m)
medC1 = global_list1['medianC'].tolist()
medA1 = global_list1['medianA'].tolist()
labels1 = global_list1['Label'].tolist()
pointsC1 = np.column_stack((medC1, labels1))
unique_pointsC1, countsC1 = np.unique(pointsC1, axis=0, return_counts=True)
pointsA1 = np.column_stack((medA1, labels1))
unique_pointsA1, countsA1 = np.unique(pointsA1, axis=0, return_counts=True)
xC1 = unique_pointsC1[:, 0]
yC1 = unique_pointsC1[:, 1]
sizesC1 = countsC1 * 120  
xA1 = unique_pointsA1[:, 0]
yA1 = unique_pointsA1[:, 1]
sizesA1 = countsA1 * 120  

# Charger les données du deuxième fichier pickle
with open(f"/home/stagiaire/D/R/metrics/MA{it}4.pkl", 'rb') as f:
    global_list2 = pickle.load(f)

# Préparer les données pour le second ensemble
data2 = {}
for i in range(1, 5):
    root_dir = f"/home/stagiaire/D/D/patchs/{i}"
    keys = {"favelas": 1, "residential": 0}
    for folder in os.listdir(root_dir):
        root_folder = os.path.join(root_dir, folder)
        files = os.listdir(root_folder)
        for file in files:
            n = file.split('.')[0]
            data2[n] = keys[folder]
labels_list2 = pd.DataFrame(list(data2.items()), columns=['Id', 'Label'])
labels_list2 = labels_list2.set_index('Id')
global_list2 = pd.merge(global_list2, labels_list2, left_index=True, right_index=True)

for index, row in global_list2.iterrows():
    if not row['PredictC']:
        global_list2 = global_list2.drop(index)
global_list2['medianC'] = (global_list2['PredictC'].apply(statistics.median)).apply(lambda m: 1 if m == 0.5 else m)
global_list2['medianA'] = (global_list2['PredictA'].apply(statistics.median)).apply(lambda m: 1 if m == 0.5 else m)
medC2 = global_list2['medianC'].tolist()
medA2 = global_list2['medianA'].tolist()
labels2 = global_list2['Label'].tolist()
pointsC2 = np.column_stack((medC2, labels2))
unique_pointsC2, countsC2 = np.unique(pointsC2, axis=0, return_counts=True)
pointsA2 = np.column_stack((medA2, labels2))
unique_pointsA2, countsA2 = np.unique(pointsA2, axis=0, return_counts=True)
xC2 = unique_pointsC2[:, 0]
yC2 = unique_pointsC2[:, 1]
sizesC2 = countsC2 * 120  
xA2 = unique_pointsA2[:, 0]
yA2 = unique_pointsA2[:, 1]
sizesA2 = countsA2 * 120  


fig, axs = plt.subplots(1, 2, figsize=(6, 3))

axs[0].scatter(xC1, yC1, s=sizesC1, color='#1f77b4', alpha=1, edgecolors="#1f77b4", label='C')
axs[0].scatter(xA1, yA1, s=sizesA1, color='#2ca02c', facecolors='none', edgecolors="#2ca02c", label='A', linewidths=1.5)
axs[0].set_xticks([0, 1])
axs[0].set_yticks([0, 1])
axs[0].set_xlim(0, 1)
axs[0].set_ylim(0, 1)
axs[0].set_xlabel('P')
axs[0].set_ylabel('L')
axs[0].set_title('')

axs[1].scatter(xC2, yC2, s=sizesC2, color='#1f77b4', alpha=1, edgecolors="#1f77b4", label='C')
axs[1].scatter(xA2, yA2, s=sizesA2, color='#2ca02c', facecolors='none', edgecolors="#2ca02c", label='A', linewidths=1.5)
axs[1].set_xticks([0, 1])
axs[1].set_yticks([0, 1])
axs[1].set_xlim(0, 1)
axs[1].set_ylim(0, 1)
axs[1].set_xlabel('P')
axs[1].set_ylabel('L')
axs[1].set_title('')

plt.tight_layout()
plt.savefig("comp.png", dpi=300, bbox_inches="tight")
plt.show()
```
